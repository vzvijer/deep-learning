```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf

%matplotlib inline
```

```python
fashion_mnist = tf.keras.datasets.fashion_mnist
```

```python
(X_train, Y_train), (X_test, Y_test) = fashion_mnist.load_data()
```

```python
N, img_h, img_w = X_train.shape
print('Train dataset shape: {} x ({} x {})'.format(N, img_h, img_w))
```

```python
plt.figure(figsize=(15, 75))
for i in range(5):
    plt.subplot(1, 5, i + 1)
    plt.imshow(X_train[i], cmap='gray')
    plt.title('Label: {}'.format(Y_train[i]))

plt.show()
```

```python
pd.Series(Y_train).value_counts()
```

```python
X_train = X_train / 255.0
X_test = X_test / 255.0
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(input_shape=(img_h, img_w)),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(10, activation='softmax')
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
result = model.fit(X_train, Y_train, validation_split=0.25, epochs=10)
```

```python
plt.plot(result.history['loss'], label='loss')
plt.plot(result.history['val_loss'], label='validation loss')
plt.legend()
plt.show()
```

```python
plt.plot(result.history['accuracy'], label='accuracy')
plt.plot(result.history['val_accuracy'], label='validation accuracy')
plt.legend()
plt.show()
```

# Convolutional Neural Network

```python
X_train = np.expand_dims(X_train, -1)
X_test = np.expand_dims(X_test, -1)

X_train.shape
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Input(shape=X_train[0].shape),
    tf.keras.layers.Conv2D(32, (3, 3), strides=2, activation='relu'),
    tf.keras.layers.Conv2D(64, (3, 3), strides=2, activation='relu'),
    tf.keras.layers.Conv2D(128, (3, 3), strides=2, activation='relu'),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(10, activation='softmax')
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
result = model.fit(X_train, Y_train, validation_split=0.25, epochs=10)
```

```python
model.evaluate(X_test, Y_test)
```
