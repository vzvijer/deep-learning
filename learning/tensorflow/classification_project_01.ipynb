```python
sns.countplot(data['loan_status'])
plt.show()
```

```python
print('In the whole dataset of {} instances, there are {} fully paid loans and {} charged off loans.'.
      format(data.shape[0], sum(data['loan_status'] == 'Fully Paid'), sum(data['loan_status'] == 'Charged Off')))
```

```python
fully_paid = data[data['loan_status'] == 'Fully Paid']
charged_off = data[data['loan_status'] == 'Charged Off']

#plt.subplot(1, 2, 1)
plt.hist(fully_paid['loan_amnt'], bins=20, color='green', alpha=0.5)
#plt.subplot(1, 2, 2)
plt.hist(charged_off['loan_amnt'], bins=20, color='red', alpha=0.5)
plt.show()
```

```python
plt.figure(figsize=(12, 10))
sns.heatmap(data.corr(), annot=True)
plt.show()
```

```python
sns.boxplot(x='loan_status', y='loan_amnt', data=data)
plt.show()
```

### term

```python
def term_to_years(term):
    if term == ' 36 months':
        return 3
    elif term == ' 60 months':
        return 5
    else:
        return 0

data['term_years'] = data['term'].apply(term_to_years)
```

```python
sns.distplot(fully_paid['loan_amnt'], color='green')
sns.distplot(data['loan_amnt'], color='red')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='sub_grade', data=data, order=sorted(data['sub_grade'].unique()), hue='loan_status')
plt.show()
```

```python
# a similar plot but it uses matplotlib
fully_paid_by_subgrade = fully_paid.groupby('sub_grade')['loan_status'].count()
charged_off_by_subgrade = charged_off.groupby('sub_grade')['loan_status'].count()

plt.figure(figsize=(10, 6))
plt.bar(fully_paid_by_subgrade.index, fully_paid_by_subgrade.values, color='blue')
plt.bar(charged_off_by_subgrade.index, charged_off_by_subgrade.values, color='orange')
plt.show()
```
