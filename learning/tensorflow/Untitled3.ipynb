```python
sns.countplot(data['loan_status'])
plt.show()
```

```python
print('In the whole dataset of {} instances, there are {} fully paid loans and {} charged off loans.'.
      format(data.shape[0], sum(data['loan_status'] == 'Fully Paid'), sum(data['loan_status'] == 'Charged Off')))
```

```python
fully_paid = data[data['loan_status'] == 'Fully Paid']
charged_off = data[data['loan_status'] == 'Charged Off']

#plt.subplot(1, 2, 1)
plt.hist(fully_paid['loan_amnt'], bins=20, color='green', alpha=0.5)
#plt.subplot(1, 2, 2)
plt.hist(charged_off['loan_amnt'], bins=20, color='red', alpha=0.5)
plt.show()
```
