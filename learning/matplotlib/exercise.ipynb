```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
diamonds = pd.read_csv('data/diamonds.csv')
diamonds.head()
```

```python
plt.scatter(diamonds['carat'], diamonds['price'], alpha=0.2, edgecolor='blue')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
plt.hist(diamonds['price'], bins=50, alpha=0.5)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
diamond_cuts = diamonds['cut'].value_counts()
plt.bar(diamond_cuts.index, diamond_cuts.values, color=['brown','pink','red','green','blue'])
plt.show()
```
